## Исключения (Exceptions) в Python 

При работе с реальными данными и реальными пользователями ошибки неизбежны.  Механизм обработки исключений в Python позволяет корректно управлять такими ошибками, предотвращая аварийное завершение программы и обеспечивая информативное сообщение для пользователя. Важной частью эффективной обработки ошибок является понимание и корректная обработка различных типов исключений.

Исключения — это события, которые нарушают нормальное выполнение программы. Каждое исключение имеет определённый *тип*, указывающий на характер ошибки. Знание распространённых типов исключений необходимо для написания точных и эффективных блоков обработки ошибок. Ниже приведены основные типы:

*   **`ValueError`:** Возникает, когда функция получает аргумент правильного типа, но с некорректным значением (например, попытка преобразовать строку "abc" в целое число).
*   **`TypeError`:** Возникает при применении операции или функции к объекту неподходящего типа (например, попытка сложить строку и число).
*   **`ZeroDivisionError`:** Возникает при попытке деления на ноль.
*   **`IndexError`:** Возникает при попытке обратиться к индексу в списке, который выходит за допустимые границы.
*   **`KeyError`:** Возникает при попытке получить доступ к несуществующему ключу в словаре.
*   **`FileNotFoundError`:** Возникает, когда Вы пытаетесь открыть файл, который не существует.
*   **`IOError`:** Общее исключение, связанное с операциями ввода-вывода. Перехватывает большинство ошибок, связанных с работой с файлами.

Попробуйте вызвать первые три исключения в ячейке ниже с помощью разного ввода:

````python
number = int(input("Ваш ввод: ")) #Добейтесь ValueError, TypeError и ZeroDivisionError
print( 10 / number)
print("10" + number)
````

## Блок `try...except` с указанием конкретных типов исключений

Чтобы исключение не прерывало выполнение программы, его можно обработать. Блок `try...except` позволяет выполнить операцию, которая может вызвать исключение, и корректно обработать это исключение, если оно возникнет. Вы можете использовать несколько блоков `except`, чтобы по-разному обрабатывать разные типы исключений.

Базовый синтаксис:

```python
try:
    # Код, который может вызвать исключение
except ExceptionType1:
     # Код для обработки исключения ExceptionType1
except ExceptionType2:
     # Код для обработки исключения ExceptionType2
...
else:
     # Код, который выполняется, если исключение не возникло
finally:
     # Код, который выполнится в любом случае (операции очистки ОП)
```

*   `try`: блок кода, в котором может возникнуть исключение.
*   `except`: один или несколько блоков кода, обрабатывающих определённые типы исключений. Если в блоке `try` произошло исключение, Python ищет соответствующий блок `except`.
*   `else`: необязательный блок, который выполняется только в случае, если в блоке `try` не было исключений.
*   `finally`: необязательный блок, который выполняется в любом случае, независимо от того, произошло исключение или нет. Обычно используется для операций очистки оперативной памяти(например, закрытие файлов, освобождение ресурсов).

Снова попробуйте разный ввод в примере ниже:

````python
try:
    number = int(input("Ваш ввод: "))
    print(f"10 / {number} = {10 / number}")
except ValueError:
    print("Вы ввели не число")
except ZeroDivisionError:
    print("На ноль делить нельзя!")
else:
    print("Исключений не возникло")
finally:
    print("Программа завершена")
````

## Класс `Exception`, создание пользовательских исключений и оператор `raise`.

Иногда возникает необходимость сигнализировать о специфических ошибках, которые не охватываются стандартными исключениями Python.

Вы можете генерировать пользовательские исключения, создав новый класс исключений и используя оператор `raise`. 

Новый класс исключений создается путём **наследования** от базового класса `Exception`. Класс-родитель указывается в скобках после команды `class` и названия класса:

```python
class MyException(Exception):
    """Новое исключение"""
    pass
...
if <условие для исключения>:
    raise MyException("Что-то не так!") # в качестве аргументов можно указать сообщение рядом с названием исключения
```

Создание пользовательских исключений позволяет предоставлять более информативные сообщения об ошибках и эффективнее обрабатывать особые ситуации.

````python
# Пример: генерация пользовательского исключения — ошибка MassValue

class InvalidMassError(Exception):
    """Эта функция генерирует пользовательское исключение, возникающее при некорректной массе звезды."""
    pass

def validate_star_mass(mass):
    """Эта функция проверяет, находится ли масса звезды в допустимом диапазоне.

    Аргументы:
        mass (float): масса звезды в массах Солнца.

    Вызывает:
        InvalidMassError: Если масса находится вне диапазона допустимых значений (от 0.08 до 100 масс Солнца).

    Возвращает:
        float: Допустимое значение массы.
    """
    if not 0.08 <= mass <= 100:
        raise InvalidMassError("Ошибка: Масса звезды должна лежать в диапазоне от 0.08 до 100 масс Солнца")
    return mass

star_mass = 1000.0
validated_mass = validate_star_mass(star_mass)
print(f"Допустимое значение массы звезды: {validated_mass} масс Солнца")
````

Для взаимодействия с пользователем мы уже использовали стандартный ввод через функцию `input()` и стандартный ввод через `print()`Файл — это именованная область данных на твердотельном накопителе (или жестком диске) компьютера. Данные в нём хранятся как `последовательность байтов`. Текстовый файл при этом можно интерпретировать как последовательность символов, если при этом известна его `кодировка` (зачастую это UTF-8, но в Windows до сих пор используется `CP1251`)

## Функция `open()` и её аргументы

Ключевой для работы с файлами является функция `open()`, возвращающая **файловый объект** (файловый поток, или дескриптор файла), являющийся ссылкой на открытый файл:

```python
file_handle = open(filename, mode='r', encoding=None)
```
Она принимает два основных аргумента:

1.  **`filename`:** имя файла — путь к файлу, который Вы хотите открыть. Это может быть относительный путь (относительно текущей рабочей директории) или абсолютный путь. Например:

    *   `"my_data.txt"` (относительный путь: файл лежит в той же папке, что и ваш код на Python)
    *   `"data/star_catalog.csv"` (относительный путь: файл находится во вложенной папке с именем "data")
    *   `"/home/user/documents/observational_data.fits"` (абсолютный путь: указывает точное расположение файла в системе; *пример для систем на базе nix*)
    *   `"C:\\Users\\MyUser\\Documents\\data.txt"` (абсолютный путь в операционной системе Windows).

2.  **`mode`:** режим — строка, определяющая, как Вы собираетесь использовать файл. Наиболее часто используемые режимы:

    * `"r"`: **режим чтения**. Открывает файл только для чтения. Это режим по умолчанию, если вы не указали другой. Вы можете только читать данные из файла. Если файл не существует, будет вызвано исключение `FileNotFoundError`.
    * `"w"`: **режим записи**. Открывает файл для записи. Если файл существует, его содержимое будет *перезаписано* (все данные будут потеряны!). Если файла не существует, он будет создан. Используйте этот режим с осторожностью.
    * `"a"`: **режим дополнения**. Открывает файл для записи, но новая информация добавляется в *конец* файла. Если файл не существует, он будет создан. Этот режим удобен, когда нужно добавить данные в существующий файл, не удаляя его содержимое.
    * `"r+"`: **режим чтения и записи**, без удаления содержимого. Файл должен существовать. Запись идёт поверх существующего текста, начиная с начала файла.
    * `'w+'`: **режим чтения и записи**, с удалением содержимого.
Файл очищается или создаётся заново. Подходит, если вы хотите сначала записать, а потом прочитать.
    * `'a+'`: **режим чтения и добавления**. Файл открывается в режиме добавления, но можно также его читать. При записи — курсор всегда в конце.
    *  `"b"`: **бинарный режим**. Открывает файл в бинарном виде (для работы с изображениями, аудио и другими бинарными данными).

Третий, опциональный аргумент `encoding`, задает кодировку файла для текстового чтения. Всегда указывайте `encoding="UTF-8"`, если вы хотите, чтобы ваша программа запускалась кроссплатформенно.

После запуска функция `open()` возвращает соответствующий файловый объект, который является итератором. Содержимое файла можно считать, обращаясь к этому итератору, например, в цикле.

Давайте, например, считаем файл `input.txt`, лежащий в каталоге с ноутбуками. Его содержимое:

```
Привет!
Это пример текстового файла.
А это - последняя строка, которая заканчивается переходом на новую.
```

````python
file_handle = open('input.txt', 'r', encoding="UTF-8")
for line in file_handle:
    print(line)
file_handle.close()
````

## Методы файловых потоков `.close()`, `.read()` и `readlines()`

После работы с файлом дескриптор необходимо закрыть с помощью метода `.close()`. Если этого не сделать, файл может остаться открытым:
* он будет заблокирован для других программ;
* данные могут не сохраниться сразу;
* при ошибке в коде `close()` может не выполниться вовсе.

Возникли лишние пустые строки — это из-за аргумента `end='\n'` у `print()` по-умолчанию. Все символы в файле читаются **как есть**, поэтому символы перехода на новую строку тоже оказываются в `line`.

Это можно увидеть, воспользовавшись чтением всего файла в одну строку через метод файлового дескриптора `.read(n=-1)`, где `n` — количество считываемых символов. Если не указать `n`, считается весь файл.

````python
file_handle = open('input.txt', 'r', encoding="UTF-8")
s = [file_handle.read()] # обернем строку в списочные скобки, чтобы явно увидеть символы перехода
print(s)
file_handle.close()
````

Сразу в список строк текстовый файл можно считать методом `.readlines()`:

````python
file_handle = open('input.txt', 'r', encoding="UTF-8")
s = file_handle.readlines() # обернем строку в списочные скобки, чтобы явно увидеть символы перехода
print(s)
file_handle.close()
````

Обертка в `try...except` на случай чтения разных файлов выполняется следующим образом. Попробуйте ввести имя файла, присутствующего в каталоге `intro`, и какое-нибудь несуществующее:

````python
filename = input("Введите имя файла: ")
try:
    file_handle = open(filename, "r", encoding='UTF-8')
    print("Файл успешно открыт!")
    file_handle.close()
    print("Файл закрыт.")

except FileNotFoundError:
    print(f"Ошибка: Файл {filename} не был найден.")
````

## Методы записи `write()` и `writelines()`

Чтобы записать данные в файл, Вам нужно открыть его либо в режиме `"w"` (запись), либо в режиме `"a"` (дополнение).

*   `file_handle.write(string)`: Записывает строку `string` в файл. Обратите внимание на то, что `write()` не добавляет автоматически символ новой строки `\n`, и стоит включить его явно самостоятельно.
*   `file_handle.writelines(list_of_strings)`: Записывает список строк в файл, тоже без символа новой строки по-умолчанию.

Запишем для примера две таблицы: разделенный запятыми CSV `catalog_1.csv` и моноширинный `catalog_2.txt`:

````python
filename = 'catalog_1.csv'

file_handle = open(filename, "w", encoding='UTF-8') 
file_handle.write("Имя объекта, Спектральный класс, Звездная величина\n")
file_handle.write("α Cen A, G2V, 0.01\n")
file_handle.write("α CMaj A, A1V, -1.46\n")
file_handle.close()
print(f"Файл {filename} создан, данные записаны.")

template = "{:>10} | {:>3} | {:>5}\n" # шаблон под моноширинную ascii-таблицу
strlist = [template.format('name', 'SpC', 'mag'),
           template.format("α Cen A", "G2V", "0.01"),
           template.format("α Cmaj A", "A1V", "-1.46")]

filename = 'catalog_2.txt'

file_handle = open(filename, "w", encoding='UTF-8') 
file_handle.writelines(strlist)
file_handle.close()
print(f"Файл {filename} создан, данные записаны.")
````

## Контекстный менеджер `with...as` и его использование вместо `close()`

После работы с файлом его необходимо закрыть. Это особенно важно, потому что:

* незакрытый файл может остаться заблокированным для других программ;
* данные из буфера могут не записаться полностью.

Однако, использование метода `close()` не всегда решает данные проблемы, ведь при возникновении ошибки до завершения работы с файлом, он останется открытым.

Более безопасный (и рекомендуемый) подход к чтению и записи файлов - использование **контекстного менеджера** `with open(...) as f`. Он гарантирует, что файл будет закрыт автоматически.

````python
with open("catalog_2.txt", encoding="UTF-8") as file_in:
    for line in file_in:
        print(line.rstrip("\n"))
    # Здесь файл автоматически закрывается
# print(file_in.readline()) # эта строка выведет ошибку ValueError
````

````python
# просто скопируем `catalog_2.txt` в `catalog_3.txt`. Контекстный менеджер можно засунуть один в другой
with open("catalog_2.txt", encoding="UTF-8") as file_in: 
    with open("catalog_3.txt", 'w', encoding="UTF-8") as file_out:
        file_out.write(file_in.read())
# При выполнении кода вне блока контекстного менеджера файлы закроются автоматически
````

## Использование print() для записи в файл

Наконец, нашего старого знакомого, функцию `print()` тоже можно использовать для вывода данных в файл.

Для этого нужно передать ей в аргумент `file` файловый объект:

````python
template = "{:>10} | {:>3} | {:>5}" # шаблон под моноширинную ascii-таблицу
strlist = [template.format('name', 'SpC', 'mag'),
           template.format("α Cen A", "G2V", "0.01"),
           template.format("α Cmaj A", "A1V", "-1.46")]

for s in strlist:
    print(s) # вывод на стандартный вывод

with open("test.txt", 'w', encoding='UTF-8') as f_out:
    for s in strlist:
        print(s, file=f_out) # вывод в файловый объект f_out

with open("test.txt", 'r', encoding='UTF-8') as f_in:
    print("Содержимое text.txt: ")
    print(f_in.read())
````
